## Neural Network model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
```

## Importing the data set

```python
train = pd.read_csv('/kaggle/input/forest-cover-type-prediction/train.csv')
test = pd.read_csv('/kaggle/input/forest-cover-type-prediction/test.csv')
```

## Prepare test and training data

Using the full feature set.

```python
X_train = train.drop(columns = ['Id', 'Cover_Type']) 
X_test = test.drop(columns = ['Id'])
y_train = train['Cover_Type'] 

print("Training features shape:", X_train.shape)
print("Training label shape:", y_train.shape)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
```

```python
y_train_encoded = to_categorical(y_train - 1)  # Subtracting 1 to convert class labels from 1 to 7 to 0 to 6
```

```python
print('Feature Array Shape:   ', X_train.shape)
print('Label Array Shape:     ', y_train.shape)
```

# Neural Network model

```python
nn = tf.keras.models.Sequential()
```

```python
nn.add(tf.keras.layers.Dense(units=50, activation='relu'))
```

```python
nn.add(tf.keras.layers.Dense(units=50, activation='relu'))
```

```python
nn.add(tf.keras.layers.Dense(units=7, activation='softmax'))
```

```python
nn.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
history = nn.fit(X_train, y_train_encoded, batch_size = 32, epochs = 100)
```

```python
# Plot training & validation accuracy values
plt.figure(figsize = (12, 6))
plt.plot(history.history['accuracy'])
plt.title('Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
```

# Submission

```python
scaler = StandardScaler()
scaler.fit(X_test) 

# Apply the scaling to the test set features
X_test_scaled = scaler.transform(X_test.values)

# Make predictions
predictions = nn.predict(X_test_scaled)

# Add 1 to convert back to class labels starting from 1
predicted_labels = np.argmax(predictions, axis = 1) + 1  
```

```python
df = pd.DataFrame({'Id': test['Id'], 'Cover_Type': predicted_labels})
df.to_csv('submission.csv', index = None)
```
